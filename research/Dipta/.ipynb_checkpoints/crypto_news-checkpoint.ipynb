```python
from openbb_terminal.sdk import openbb
# Show trending coins
openbb.crypto.disc.trending()
```

```python
df_news = bb.crypto.ov.news(post_kind = 'news',region = 'en', 
                  filter_= 'hot', ascend = False)
```

```python
df_news.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'published_at' and calculate the mean for negative and positive votes.
daily_avg_votes = df_news.groupby('published_at').agg({
    'negative_votes': 'mean',
    'positive_votes': 'mean'
}).reset_index()

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(daily_avg_votes['published_at'], daily_avg_votes['negative_votes'], label='Avg Negative Votes', marker='o', color='red')
plt.plot(daily_avg_votes['published_at'], daily_avg_votes['positive_votes'], label='Avg Positive Votes', marker='o', color='green')

plt.title("Average Votes Over Time")
plt.xlabel("Publication Date")
plt.ylabel("Average Number of Votes")
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better visibility

plt.show()
```

```python
daily_votes_stats = df_news.groupby('published_at').agg({
    'negative_votes': ['mean', 'median'],
    'positive_votes': ['mean', 'median']
}).reset_index()

# Flatten the MultiIndex for easier plotting.
daily_votes_stats.columns = ['_'.join(col).rstrip('_') for col in daily_votes_stats.columns.values]

# Plotting
plt.figure(figsize=(12, 6))

# Average values
plt.plot(daily_votes_stats['published_at'], daily_votes_stats['negative_votes_mean'], label='Avg Negative Votes', marker='o', color='red', linestyle='--')
plt.plot(daily_votes_stats['published_at'], daily_votes_stats['positive_votes_mean'], label='Avg Positive Votes', marker='o', color='green', linestyle='--')

# Median values
plt.plot(daily_votes_stats['published_at'], daily_votes_stats['negative_votes_median'], label='Median Negative Votes', marker='x', color='red')
plt.plot(daily_votes_stats['published_at'], daily_votes_stats['positive_votes_median'], label='Median Positive Votes', marker='x', color='green')

plt.title("Average and Median Votes Over Time")
plt.xlabel("Publication Date")
plt.ylabel("Number of Votes")
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better visibility

plt.show()
```
